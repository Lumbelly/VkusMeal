Создание дополнительного столбца для создания категорий для сетов и допродаж

```python
# Создаем пустой столбец 'category'
data['category'] = ''

# Словарь со значениями для категории "сеты"
sets_keywords = [
    'соки из вкусвилл', 
    'смузи из вкусвилл', 
    'фруктовые нарезки', 
    'овощные нарезки', 
    'десерт на каждый день', 
    'круассан для настроения', 
    'чиабатта к супу'
]

# Присваиваем категорию "сеты"
data.loc[data['name'].str.contains('|'.join(sets_keywords), case=False, na=False), 'category'] = 'сеты'

# Список ключевых слов для категории "допродажи"
upsell_keywords = [
    'на яблочном соке', 
    'конфета',
    'чипсы', 
    'снеки', 
    'манго-апельсин', 
    'напиток',
    'снек',
    'морская', 
]

# Присваиваем категорию "допродажи"
data.loc[data['name'].str.contains('|'.join(upsell_keywords), case=False, na=False), 'category'] = 'допродажи'

# Присваиваем категорию "рационы" для всех записей, которые еще не имеют категории
data.loc[data['category'] == '', 'category'] = 'рационы'

# Выводим итоговый DataFrame
data.head()
```

```python
data.groupby('city')['item_id'].count()
```

Переименование рационов в единый формат

```python
# Определяем функцию для присвоения значений в столбец 'name'
def determine_ration(name):
    name = name.lower()  # Приводим к нижнему регистру для корректного сравнения
    if 'зц' in name:
        return 'зц'
    elif 'комбо' in name:
        return 'комбо'
    elif 'комб' in name:
        return 'комбо'
    elif 'баланс' in name:
        return 'баланс'
    elif 'ввг' in name or 'готовит' in name:
        return 'ввг'
    elif 'базовый' in name:
        return 'базовый'
    elif 'семейный' in name:
        return 'семейный'
    elif 'супер' in name:
        return 'супер'
    else:
        return name  # Возвращаем исходное значение, если ни одно условие не подошло

# Применяем функцию к столбцу 'name' и перезаписываем его
data['name'] = data['name'].apply(determine_ration)

# Выводим обновленный DataFrame
data.groupby('name')['item_id'].count()
```

### Расчет распределения сетов и доптоваров для каждого рациона. Поиск лидеров

Добавяем датафрейм где заказов больше 2 на одну дату

```python
# Группировка по 'order_id' и 'order_date', подсчет числа заказов
grouped = data.groupby(['order_id', 'order_date']).size().reset_index(name='count')

# Фильтрация групп с количеством заказов 2 и более
filtered_orders = grouped[grouped['count'] >= 2]

# Слияние исходного DataFrame с отфильтрованными заказами для получения полной информации
result_df = data.merge(filtered_orders[['order_id', 'order_date']], on=['order_id', 'order_date'])

# Вывод результата
result_df
```

Оставляем только комбинации с рационами

```python
# Группировка и фильтрация
filtered = (result_df.groupby(['order_id', 'order_date'])['category']
            .nunique()
            .reset_index(name='unique_categories_count'))

# Оставляем только те записи, где количество уникальных категорий больше 1
filtered = filtered[filtered['unique_categories_count'] > 1]

# Объединение с исходным DataFrame для получения нужных строк
final_result = pd.merge(result_df, filtered[['order_id', 'order_date']], 
                         on=['order_id', 'order_date'])

# Вывод финального результата
final_result.head(10)
```

Найдем топ комбинаций для Москвы

```python
# Шаг 1: Фильтруем данные для Москвы
msc_combinations = final_result.query('city == "moscow"')

# Шаг 2: Группируем данные и собираем все name в списки
msc_combinations = msc_combinations.groupby(['order_id', 'order_date'])['name'].apply(list).reset_index()

# Шаг 3: Функция для подсчета комбинаций
def count_combinations(names):
    combs = combinations(sorted(set(names)), 2)  # Генерируем все возможные комбинации 2-х элементов
    return Counter(combs)

# Шаг 4: Применяем функцию к каждой группе и считаем комбинации
msc_combinations['combinations'] = msc_combinations['name'].apply(count_combinations)

# Шаг 5: Объединяем результаты в один общий счетчик
total_combinations = Counter()
for count in msc_combinations['combinations']:
    total_combinations.update(count)

# Шаг 6: Преобразуем счетчик в DataFrame для удобства
msc_combinations_data = pd.DataFrame(total_combinations.items(), columns=['combination', 'count'])

# Выводим итоговые комбинации, отсортированные по убыванию
result = msc_combinations_data.sort_values(by='count', ascending=False)

# Фильтруем топ-5 комбинаций
top_5_combinations = result.head(5)

# Извлекаем данные для диаграммы
labels = top_5_combinations['combination'].astype(str)  # Преобразуем комбинации в строку для меток
sizes = top_5_combinations['count']  # Количество для каждой комбинации

# Настройка диаграммы пирога
plt.figure(figsize=(10, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'), explode=[0.1]*len(labels))

# Заголовок диаграммы
plt.title('Топ-5 комбинаций пользователей в Москве', fontsize=16)
plt.axis('equal')  # Гарантирует, что пирог будет круговым

# Отображение диаграммы
plt.show()
```
