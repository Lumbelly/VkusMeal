```python
data['sale_price_1_day'] = ""
```

```python
def bazovy_data(data, target_value):
    for index, row in data.iterrows():
        if target_value in row['product']:
            if row['days'].isnumeric() and int(row['days']) <= 14:  # Проверяем, что 'days' является числом
                data.loc[index, 'sale_price_1_day'] = row['price_1_day']
            else:
                data.loc[index, 'sale_price_1_day'] = row['price_1_day'] * 0.95  # Вычитаем 5%

    return data

target_values = ['Базовый 3 блюда', 'Базовый 4 блюда', 'Базовый 5 блюд']
for target_value in target_values:
    data = bazovy_data(data, target_value)

filtered_data = data[data['product'].str.contains('Базовый')]
data[data['product'].str.contains('Базовый 5 блюд')]
```

```python
def ostalnie_data(data, target_value):
    for index, row in data.iterrows():
        if target_value in row['product']:
            try:
                days = int(row['days'])
                if days <= 7:
                    data.loc[index, 'sale_price_1_day'] = row['price_1_day']
                elif days <= 14:
                    data.loc[index, 'sale_price_1_day'] = row['price_1_day'] * 0.98  # Вычитаем 2%
                elif days <= 28:
                    data.loc[index, 'sale_price_1_day'] = row['price_1_day'] * 0.95  # Вычитаем 5%
                else:
                    data.loc[index, 'sale_price_1_day'] = row['price_1_day'] * 0.90  # Вычитаем 10%
            except ValueError:
                # Пропуск строки или принятие альтернативного решения
                pass

    return data

target_values = ['Баланс 1200', 'Баланс 1500', 'Баланс 1800', 'Баланс 2100', 'Баланс 2400', 'Баланс 850', 'ВВ Готовит 3 блюда', \
                 'ВВ Готовит 4 блюда', 'ВВ Готовит 5 блюд', 'ВВ Готовит 6 блюд', 'ВВ Готовит 7 блюд', 'Гурмэ 3 приема', \
                 'Гурмэ 4 приема', 'Гурмэ 5 блюд', 'Гурмэ 6 блюд', 'Комбо Баланс 1200+1500', 'Комбо Баланс 850+1200', \
                 'Комбо Баланс 850+850', 'Комбо ВВГ 3+3 блюда', 'Комбо ВВГ 3+4 блюда', 'Комбо ВВГ 4+4 блюда', 'Комбо ВВГ 4+5 блюд', \
                 'Комбо ВВГ 5+5 блюд', 'Семейный 6 блюд+фрукт', 'Семейный 6Блюд', 'Семейный 8 блюд+фрукт', 'Семейный 8Блюд']

for target_value in target_values:
    data = ostalnie_data(data, target_value)

filtered_data = data[~data['product'].str.contains('Базовый')]
data[data['product'].str.contains('Базовый 5 блюд')]
```

```python
# Заменить пустые строки в столбце 'sale_price_1_day' на NaN
data['sale_price_1_day'] = data['sale_price_1_day'].replace('', np.nan)

# Заполнить пропущенные значения в столбце 'sale_price_1_day' нулями (или другим нужным значением)
data['sale_price_1_day'] = data['sale_price_1_day'].fillna(0)

# Преобразовать столбец 'sale_price_1_day' в целочисленный формат
data['sale_price_1_day'] = data['sale_price_1_day'].astype(int)
```

```python
# Заменить пустые строки в столбце 'sale_price_1_day' на NaN
data['price_1_day'] = data['price_1_day'].replace('', np.nan)

# Заполнить пропущенные значения в столбце 'sale_price_1_day' нулями (или другим нужным значением)
data['price_1_day'] = data['price_1_day'].fillna(0)

# Преобразовать столбец 'sale_price_1_day' в целочисленный формат
data['price_1_day'] = data['price_1_day'].astype(int)
```

```python
# Заменить пустые строки в столбце 'sale_price_1_day' на NaN
data['days'] = data['days'].replace('', np.nan)

# Заполнить пропущенные значения в столбце 'sale_price_1_day' нулями (или другим нужным значением)
data['days'] = data['days'].fillna(0)

# Преобразовать столбец 'sale_price_1_day' в целочисленный формат
data['days'] = data['days'].astype(int)
```

```python
data.info()
```

```python
# Выполнение операции вычитания после преобразования данных
data['sale_1_day'] = data['price_1_day'] - data['sale_price_1_day']
data[data['product'].str.contains('Базовый 5 блюд')]
```

```python
data['revenue'] = data['price_1_day'] * data['days']
data['revenue_net'] = data['sale_price_1_day'] * data['days']
data['sale'] = data['revenue'] - data['revenue_net'] #сикдка за количество дней для одной позиции

data[data['product'].str.contains('Базовый 5 блюд')]
```

Считаем июнь

```python
june = data.query('month == 6')
```

```python
june_data = june.groupby(['product', 'days'])[['revenue', 'sale']].sum().reset_index()
june_data['count'] = june.groupby(['product', 'days']).size().reset_index(name='count')['count']
june_data['percent_sale'] = (june_data['sale'] / june_data['revenue']) * 100

june_data
```

```python
june_pivot = june_data.pivot_table(index=['product', 'days'], values=['sale', 'count'], aggfunc='sum').reset_index().sort_values(by=['product', 'days'])
june_final_data = june_pivot
june_final_data[june_final_data['product'].str.contains('Базовый 5 блюд')]
```
